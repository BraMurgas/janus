# Example of application of the discrete Green operator (using serial FFTs)

```python
%pylab --no-import-all inline

# Update sys.path in order to make locally installed modules available
import sys
sys.path.append('./src/')

import fft.serial
import greenop
import discretegreenop
import numpy.fft as npfft

from matprop import IsotropicLinearElasticMaterial as Material
```

```python
def my_imshow(a):
    plt.imshow(a, cmap=plt.cm.Greys_r, interpolation='nearest')
```

```python
dim = 2
sym = (dim * (dim + 1)) // 2

nx = 128
ny = 64

mu = 0.75
nu = 0.3

# Size of each voxel (unimportant)
h = 1.0
```

```python
mat = Material(mu, nu, dim)
green = discretegreenop.TruncatedGreenOperator(greenop.create(mat), (nx, ny), h)
```

## Generation of the reference truncated Green operator

```python
def delta(i, j):
    if i == j:
        return 1
    else:
        return 0
```

```python
def gamma(kx, ky, i, j, h, l):
    n = np.empty((2,), dtype=np.float64)
    norm = np.sqrt(kx * kx + ky * ky)
    if norm == 0:
        return 0.
    else:
        n[0] = kx / norm
        n[1] = ky / norm
        return ((delta(i, h) * n[j] * n[l]
                 + delta(i, l) * n[j] * n[h]
                 + delta(j, h) * n[i] * n[l]
                 + delta(j, l) * n[i] * n[h]) / (4 * mu)
                - 1 / (2 * mu * (1 - nu)) * n[i] * n[j] * n[h] * n[l])
```

```python
kx = np.arange(nx, dtype=np.float64)
index = 2 * kx > nx
kx[index] -= nx
kx *= 2. * np.pi / nx

ky = np.arange(ny, dtype=np.float64)
index = 2 * ky > ny
ky[index] -= ny
ky *= 2. * np.pi / ny
```

```python
expected_green = np.empty((ny, nx, sym, sym), dtype=np.float64)
sqrt_two = np.sqrt(2.)

for i in range(ny):
    for j in range(nx):
        expected_green[i, j, 0, 0] = gamma(kx[j], ky[i], 0, 0, 0, 0)
        expected_green[i, j, 0, 1] = gamma(kx[j], ky[i], 0, 0, 1, 1)
        expected_green[i, j, 0, 2] = sqrt_two * gamma(kx[j], ky[i], 0, 0, 0, 1)
        expected_green[i, j, 1, 0] = gamma(kx[j], ky[i], 1, 1, 0, 0)
        expected_green[i, j, 1, 1] = gamma(kx[j], ky[i], 1, 1, 1, 1)
        expected_green[i, j, 1, 2] = sqrt_two * gamma(kx[j], ky[i], 1, 1, 0, 1)
        expected_green[i, j, 2, 0] = sqrt_two * gamma(kx[j], ky[i], 0, 1, 0, 0)
        expected_green[i, j, 2, 1] = sqrt_two * gamma(kx[j], ky[i], 0, 1, 1, 1)
        expected_green[i, j, 2, 2] = 2. * gamma(kx[j], ky[i], 0, 1, 0, 1)
```

```python
# For the xxxy and yyxy coefficients, set the highest frequency to zero,
# as the symmetries for real DFT cannot be enforced there
expected_green[ny / 2, :, 0:1, 2] = 0.
expected_green[:, nx / 2, 0:1, 2] = 0.
expected_green[ny / 2, :, 2, 0:1] = 0.
expected_green[:, nx / 2, 2, 0:1] = 0.
```

```python
for i in range(sym):
    for j in range(sym):
        plt.subplot(sym, sym, sym * j + i + 1)
        my_imshow(expected_green[:, :, i, j])
        sink = plt.axis('off')
```

```python
my_imshow(expected_green[:, :, 0, 2])
```

## Application in the Fourier space

This application allows the vizualization of the components of $\Gamma_0^h$ in the Fourier space.

```python
actual_green = np.empty((ny, nx, sym, sym))
b = np.empty((2,), dtype=np.int32)
for b[0] in range(nx):
    for b[1] in range(ny):
        green.asarray(b, actual_green[b[1], b[0], :, :])
```

```python
for i in range(sym):
    for j in range(sym):
        plt.subplot(sym, sym, sym * j + i + 1)
        my_imshow(actual_green[:, :, i, j])
        sink = plt.axis('off')
```

```python
index = expected_green != 0.0
np.max(np.abs((expected_green[index] - actual_green[index]) / expected_green[index]))
```

## Application in the real space (localized polarization)

```python
transform = fft.serial.create_real((ny, nx))
```

### Case of a localized polarization: computation of the Green operator in the real space

In this application, the polarization is non-zero only in the top-left voxel, and the Green operator is applied to this unit sollicitation. This effectively computes the discrete Green operator in the real space! Application of the Green operator to *any* polarization field would reduce (in the real space) to a convolution of this computation with the polarization field under consideration.

The results are then compared with reference data, computed with a `Java` code.

```python
tau = np.zeros((ny, nx, sym), dtype=np.float64)
tau[0, 0, 0] = 1
dft_tau = np.empty(transform.cshape + (sym,), dtype=np.float64)
for i in range(sym):
    dft_tau[:, :, i] = transform.r2c(tau[:, :, i])
```

```python
actual_dft_eta = np.empty_like(dft_tau)

b = np.empty((2,), dtype=np.int)
for b[1] in range(transform.cshape[0]):
    for b[0] in range(transform.cshape[1] // 2):
        actual_dft_eta[b[1], 2 * b[0], :] = green.apply(b, dft_tau[b[1], 2 * b[0], :])
        actual_dft_eta[b[1], 2 * b[0] + 1, :] = green.apply(b, dft_tau[b[1], 2 * b[0] + 1, :])
```

```python
actual_eta = np.empty_like(tau)
for i in range(sym):
    actual_eta[:, :, i] = np.asarray(transform.c2r(actual_dft_eta[:, :, i])) / (nx * ny)
```

```python
# Generation of reference data
g = expected_green
dft_tau = np.empty(tau.shape, dtype=np.complex128)
for i in range(tau.shape[-1]):
    dft_tau[:, :, i] = npfft.fft2(tau[:, :, i])
    
expected_dft_eta = np.empty_like(dft_tau)
expected_dft_eta[:, :, 0] = g[:, :, 0, 0] * dft_tau[:, :, 0] + g[:, :, 0, 1] * dft_tau[:, :, 1] + g[:, :, 0, 2] * dft_tau[:, :, 2]
expected_dft_eta[:, :, 1] = g[:, :, 1, 0] * dft_tau[:, :, 0] + g[:, :, 1, 1] * dft_tau[:, :, 1] + g[:, :, 1, 2] * dft_tau[:, :, 2]
expected_dft_eta[:, :, 2] = g[:, :, 2, 0] * dft_tau[:, :, 0] + g[:, :, 2, 1] * dft_tau[:, :, 1] + g[:, :, 2, 2] * dft_tau[:, :, 2]

expected_eta = np.empty_like(expected_dft_eta)
for i in range(expected_eta.shape[-1]):
    expected_eta[:, :, i] = npfft.ifft2(expected_dft_eta[:, :, i])
```

```python
index = np.abs(expected_eta) > 1E-15
np.max(np.absolute((actual_eta[index] - expected_eta[index]) / expected_eta[index]))
```

```python
np.max(np.abs(expected_eta[:, :, :].imag))
```

```python
np.max(np.abs(expected_eta[:, :, 2].real - actual_eta[:, :, 2]))
```

### Case of a polarization which is translation invariant along the $y$ axis.

In this application, the polarization $\tau_{ij}$ is of the form

$$\tau_{xx}=\sin\frac{2\pi x}{L}$$

(other $\tau_{ij}=0$). The resulting strain field should also have a preferred direction along the $x$-axis. This application therefore allows to check that row and column indices are all handled correctly.

```python
dummy = np.sin(2 * np.pi * np.arange(nx) / nx)
dummy.shape=(1, nx)
tau[:, :, 0] = np.repeat(dummy, ny, axis=0)
dft_tau = np.empty(transform.cshape + (sym,), dtype=np.float64)
for i in range(sym):
    dft_tau[:, :, i] = transform.r2c(tau[:, :, i])
```

```python
plt.imshow(tau[:, :, 0], cmap=plt.cm.Greys_r, interpolation='nearest')
sink = plt.axis('off')
```

```python
dft_eta = np.empty_like(dft_tau)

b = np.empty((2,), dtype=np.int)
for b[1] in range(transform.cshape[0]):
    for b[0] in range(transform.cshape[1] // 2):
        dft_eta[b[1], 2 * b[0], :] = green.apply(b, dft_tau[b[1], 2 * b[0], :])
        dft_eta[b[1], 2 * b[0] + 1, :] = green.apply(b, dft_tau[b[1], 2 * b[0] + 1, :])
        
eta = np.empty_like(tau)
for i in range(sym):
    eta[:, :, i] = transform.c2r(dft_eta[:, :, i])
```

```python
for i in range(sym):
    plt.subplot(1, sym, i + 1)
    plt.imshow(eta[:, :, i], cmap=plt.cm.Greys_r, interpolation='nearest')
    sink = plt.axis('off')
```
